#### **__Limitations of DBSCAN__**

1. My dataset contains categorical features; clustering techniques doesn't work on categorical features.
2. It is sensitive to choice of `ε-neighbourhood` and `minimum point` required to make a point a core point. 
    - `ε` that is too small will leax to sparse clusters being classified as noise.
    - `ε` that is too large will result in the merging of dense clusters.
3. It cannot handle a dataset with varying density or different local densities.

#### **__Limitations of FAMD__**

FAMD's limitation is its treatment of categorical variables. `FAMD` treats all qualitative attributes as nominal variables and creates dummy variables for each level of these features. However, creating dummy variables for ordinal variables may be inappropriate because they have a specific order or ranking associated with their levels. Therefore, treating ordinal variables as nominal will result in losing important information or misrepresenting the underlying data structure.

#### *__Is there an association between the categorical features?__*

To answer this question, one could consider the following approaches:

1. Use Cramer's V statistics to measure the strength of association between pairs of `nominal` features.
2. Employ Kendall's Tau or Spearman rank correlation to measure the strength of association between pairs of `ordinal` features.
3. In addition, the chi-square test of independence can assess the significance of the association between any pair of categorical (nominal or ordinal) features. If the sample size is small, use the Fisher exact test of independence instead.

For this analysis, we will use the chi-square test of independence or Fisher's exact test to assess the association between the categorical features.

• In addition, the ANOVA or it’s non-parametric equivalent, Kruskal-Wallis test, will be 
used to test for the effect of each categorical variable on each of the quantitative or 
numeric variables.

```python
import scipy.stats as stats 

cols = nominal_features + ordinal_features
p_table = pd.DataFrame(index = cols, columns= cols)
np.fill_diagonal(p_table.values, 1.0)

n = len(cols)
for i in range(n-1):
    for  j in range( i+1, n):
        
        col1, col2= cols[i], cols[j]
        contingency_table = pd.crosstab(data[col1], data[col2])
        
        chi2_stat, p_value, _, _ = stats.chi2_contingency(contingency_table)
        
        p_table.loc[col1, col2] = p_value
        p_table.loc[col2, col1] = p_value

    
print("P-Value Table (Chi-Squared Test of Independence):\n")
print(p_table)
```

<span style="color:red">**Comments:** </span> Results from the chi-square test above indicates that:

There is a statistically significant association `(p-value < 0.05)` between the following features:
   - thal and all features but `restecg and fbs`
   - exang and all features but `restecg and fbs`
   - slope vs cp

Given that we have features with significant association, there is a potential problem of multicollinearity. To address this concern, we would consider doing the following for downstream analysis:

- remove one or more features that are correlated.
- use dimension reduction techniques to reduce the number of features in our dataset

#### __Dimension Reduction to Understand the data__

Given that the `heart disease` dataset includes both numerical and categorical features, typical methods like Principal Component Analysis (PCA), Singular Value Decomposition, or traditional Factor Analysis that are designed for numerical data will not be suitable. To address this challenge, we can use an extension of the traditional Factor Analysis and PCA called the Factor Analysis of Mixed Data (FAMD). FAMD is specifically designed to handle datasets containing a combination of numerical and categorical variables. It transforms these variables into a suitable format for factor analysis, enabling us to identify underlying patterns and reduce the dataset's dimensionality while retaining the information contained in both types of variables.

Before employing DBSCAN, it is important to determine the optimal `ε` and `minimum points`.  The `k-distance graph` would be used to estimate `ε`.

#### __Estimating `ε` Using k_Distance Graph__

```python
# Source: https://scikit-learn.org/stable/modules/generated/sklearn.neighbors.NearestNeighbors.html
        # https://github.com/scikit-learn/scikit-learn/blob/main/sklearn/neighbors/_unsupervised.py
        
# Plot the K-Distance Graph to choose epsilon (ε)

k = 5

neighbours = NearestNeighbors(n_neighbors=k)
nbrs = neighbours.fit(famd_df)
distances, _ = nbrs.kneighbors(famd_df)
k_distances = sorted(distances[:, -1],reverse=True)

plt.figure(figsize=(7, 7))
plt.hist(k_distances, bins= 14, color='skyblue', edgecolor='black')
plt.axvline(x=3, color='red', linestyle='--', linewidth=2)

plt.xlabel('K-distances')
plt.ylabel('Frequency')
plt.title(f'')
plt.grid(True)
plt.show()
```

Based on the distribution of distances above, our choice of `ε` is 3 and the `minimum point` is [4,5] .

#### __DBSCAN__

```python
epsilon = 3
minPts = 4

dbscan = DBSCAN(eps=epsilon, min_samples=minPts)
famd_df['dbCluster'] = dbscan.fit_predict(famd_df)

unique_clusters = famd_df['dbCluster'].value_counts()
print(unique_clusters)
```
